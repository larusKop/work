```python
import pandas as pd
from pandas import Series, DataFrame
ratings = pd.read_table('C:/Users/larus/Desktop/movielens/ratings.dat.txt', sep='::', names=['UserID', 'MovieID', 'Rating', 'Timestap'], engine='python')
users = pd.read_table('C:/Users/larus/Desktop/movielens/users.dat.txt', sep='::', names = ['UserID', 'Gender', 'Age', 'Occepation', 'Index'], engine='python')
data_ru = pd.merge(ratings, users)
movies = pd.read_table('C:/Users/larus/Desktop/movielens/movies.dat.txt', sep='::', names=['MovieID', 'Title', 'Tag'], engine='python')
data_all = pd.merge(data_ru, movies)
mean_ratings_gender = pd.pivot_table(data_all, values='Rating', index='Title', columns='Gender', aggfunc='mean', fill_value=0)
# Разная оценка: женская - высокая, мужская - низкая
mean_ratings_gender['diff'] = mean_ratings_gender['F'] -  mean_ratings_gender['M']
mean_ratings_gender.sort_values(by='diff', ascending=False).head(15)
```
