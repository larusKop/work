```python
dell=int(input("введите кол-во элементов в дочернем списке "))
a = [i for i in range(10, 111)]
b = []
index_for_my_little_mas = -1
for i in range(0, len(a) - 1):
    if i % dell == 0:
        b.append([])
        index_for_my_little_mas += 1
    b[index_for_my_little_mas].append(a[i])

for i in range(len(b)):
    print(b[i])
```
